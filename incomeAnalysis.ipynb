```python
import pandas as pd
import numpy as np
```

```python
income = pd.read_csv('data/income tentative.csv')
```

```python
cost = pd.read_csv('data/costOfLiving.csv')
```

```python
income = income[income.stateProvince != 'US']
max_income = income.max(axis = 0,numeric_only=True)
print(max_income)
```

```python
income['avgIncome'] = income['avgIncome'].apply(lambda x: x / max_income)
income
```

```python
cost = cost.groupby(['stateProvince'])['index'].mean().reset_index(name="index")
cost
```

```python
cost['index'] = cost['index'].apply(lambda x: 100/x)
cost
```

```python
income_data = income.merge(cost,how='outer')
income_data
```

```python
income_data['adjustedIncome'] = income_data['avgIncome'] * income_data['index']
income_data
```

```python
income_data.to_csv('data/income.csv')
```
