```python
import pandas as pd
import numpy as np
```

```python
pop_by_zip = pd.read_csv('data/population_by_zip_2010.csv')
```

```python
us_postal_codes = pd.read_csv('data/us_postal_codes.csv')
```

```python
popzip = pop_by_zip.groupby('zipcode')['population'].sum().reset_index(name="popsum")
popzip.head(20)
```

```python
temp = us_postal_codes.merge(popzip,how='outer',on='zipcode')
```

```python
temp = temp.dropna()
us_postal_codes = temp
us_postal_codes
```

```python
us_postal_codes.columns
```

```python
keys = ['city','state']
population = us_postal_codes.groupby(['city','state'])['popsum'].sum().reset_index(name="pop")
population.head(100)
```

```python
population.to_csv('data/population.csv')
```
