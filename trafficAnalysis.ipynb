```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/Preprocessed.csv')
```

```python
filtered_data = data.filter(['stateProvince','city','clicks','localClicks'])
filtered_data.head(20)
```

```python
click_grouped = filtered_data.groupby(['city','stateProvince'])['clicks'].sum().reset_index(name="clicks")
click_grouped.head(20)
```

```python
local_click_grouped = filtered_data.groupby(['city','stateProvince'])['localClicks'].sum().reset_index(name="localClicks")
local_click_grouped.head(20)
```

```python
all_clicks = click_grouped.merge(local_click_grouped,how='outer')
all_clicks['nonlocalClicks'] = all_clicks['clicks']-all_clicks['localClicks']
all_clicks.head(20)
```

```python
all_clicks.to_csv('data/traffic.csv')
```
