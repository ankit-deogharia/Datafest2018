```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/Preprocessed.csv')
```

```python
filtered_data = data.filter(['city','stateProvince','companyId','avgOverallRating'])
filtered_data.head(20)
```

```python
#company_count = filtered_data.groupby(['city','stateProvince'])['companyId'].count().reset_index(name="count")
#company_count.head(20)
```

```python
dropped = filtered_data
dropped.loc[dropped['avgOverallRating'] == 0.0, 'avgOverallRating'] = 2.5
dropped.head(20)
```

```python
satisfaction = dropped.groupby(['city','stateProvince'])['avgOverallRating'].mean().reset_index(name="avgOverallRating")
satisfaction.head(20)
```

```python
satisfaction['avgOverallRating'] = satisfaction['avgOverallRating'].apply(lambda x: x / 5.0)
satisfaction.head(20)
#satisfaction.shape
```

```python
satisfaction.to_csv('data/satisfaction.csv')
```
